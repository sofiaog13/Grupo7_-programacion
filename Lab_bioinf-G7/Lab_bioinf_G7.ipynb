```python
with open('/content/drive/MyDrive/Copia de sequence_Covid-19.fasta') as f: #se usa la palabra reservada with para que el programa abra y cierre el fihcero
    lines = f.readlines() #se crea una variable lines para guardar la lectura completa de lo que contiene el fichero
print(lines,"\n")#se imprime la variable lines en consola
secuence=''#se declara la variable secuence de tipo str
for l in lines[1:]:#se crea un bucle for para guardar la secuencia del archivo en un solo string y omitiendo la primera linea de texto
  secuence+=l[:-1]
print(secuence,"\n")#se imprime la variable secuence
problema='AAGGCAGATGGGCTATATAAACGTTTTCGCTT'#se declara la variable problema de tipo str para guardar la secuencia a hallar
def funcion(a,b):#se define una funcion cque recibe dos argumentos
  x=0#se declara la variable local x de tipo int
  for i in a: #se crea un bucle for que recorre la variable ingresada en el primer argumento
    if a[x:x+32]==b: #se especifica la longitud del fragmento se va a comparar (de x a x+32 porque son 31 letras)
      break #si el fragmento es igual al de la secuencia de la segunda variable el ciclo se termina
    else:
      x+=1 # se el fragmento no es igual se suma 1 a x y el ciclo se repite con el siguiente fragmento de la secuencia
  t=(x,x+31)#cuando el ciclo termina se guarda en una tupla el valor del indice de la primera y ultima letra
  return(t)#la funcion retorna la tupla 
funcion(secuence,problema) #se llama a la funcion con la variable secuence y problema como argumentos
```
