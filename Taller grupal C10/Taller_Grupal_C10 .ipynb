Cree una imagen de un canal, es decir con una forma (alto, ancho) del tamaño que desee, donde la diagonal principal hacia arriba sea blanca y desde la diagonal hacia abajo sea negra.


```python
import matplotlib.pyplot as plt #se importa la libreria 
imagen=[] #se declara la matriz vacia
for i in range(0,100): #bucle definido for para i en rango de 0 a 100 (alto, eje y)
  l=[] #se declara una lista vacia para ir añadiendo los datos
  for j in range(0,100): #bucle definido for para j en rango de 0 a 100 (ancho, eje x)
    if j<i: #si j es menor que i se designa 0(color negro) a todos los valores del pixel
     pixel=[0,0,0]
    else: #si no se cumple la condicion del if entonces se designa 255(color blanco) a todos los valores del pixel
     pixel=[255,255,255]
    l.append(pixel) #al final de cada ciclo del for anidado se añade el pixel a la lista 
  imagen.append(l)#al final de cada ciclo del for principal se añade la lista y todos los datos añadidos a la matriz "imagen"
plt.imshow(imagen) #se muestra en panatalla la imagen creada
```

Use una imagen de la librería skimage, en la base de datos “data”, e implemente alguno de los siguientes procesos:

Suavizar y resaltar contornos

Restauración

Eliminación de ruido

Mejoramiento del contraste

Detección de bordes



```python
# Importar la librerias a usar
from skimage import data  #Base de datos de donde viene la imagen
from skimage import feature #Importar rasgos
from skimage.color import rgb2gray # Transforma RGB a escala de grises
import matplotlib.pyplot as plt #biblioteca matplotlib
#se implementa el proceso de suavizar y resaltar contornos
imagen=data.astronaut() #Base de datos. y la imagen de esta en este caso astronaut
imagen_2=rgb2gray(imagen) #Aplica escala de grises a la imagen
edge=feature.canny(imagen_2) #Aplica el filtro canny
plt.imshow(edge) #mostrar datos como una imagen
plt.show()#Mostrar todas las figuras abiertas
```
